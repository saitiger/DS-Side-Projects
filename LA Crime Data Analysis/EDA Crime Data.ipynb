```python
def calculate_yoy_growth(df):
    # Get monthly counts
    monthly_counts = df[df['DATE OCC'].dt.year < 2025].groupby(df['DATE OCC'].dt.strftime('%Y-%m'))['DR_NO'].count()
    
    # Convert to DataFrame for easier manipulation
    monthly_df = monthly_counts.reset_index()
    monthly_df.columns = ['Date', 'Crime_Count']
    
    # Convert to datetime for proper sorting and year-month extraction
    monthly_df['Date'] = pd.to_datetime(monthly_df['Date'] + '-01')
    monthly_df = monthly_df.sort_values('Date')
    
    # Extract year and month for comparison
    monthly_df['Year'] = monthly_df['Date'].dt.year
    monthly_df['Month'] = monthly_df['Date'].dt.month
    monthly_df['MonthName'] = monthly_df['Date'].dt.strftime('%b')
    
    # Create a new DataFrame to store YoY growth
    yoy_growth = []
    
    # Calculate YoY growth for each month
    for year in sorted(monthly_df['Year'].unique())[1:]:  # Skip the first year as we need previous year data
        for month in range(1, 13):  # For each month
            # Current month data
            current = monthly_df[(monthly_df['Year'] == year) & (monthly_df['Month'] == month)]
            
            # Previous year's same month
            previous = monthly_df[(monthly_df['Year'] == year-1) & (monthly_df['Month'] == month)]
            
            # Only calculate if we have data for both periods
            if not current.empty and not previous.empty:
                current_count = current['Crime_Count'].values[0]
                previous_count = previous['Crime_Count'].values[0]
                
                # Calculate percentage growth
                growth_pct = ((current_count - previous_count) / previous_count) * 100
                
                month_name = current['MonthName'].values[0]
                
                yoy_growth.append({
                    'Year': year,
                    'Month': month,
                    'MonthName': month_name,
                    'CurrentCount': current_count,
                    'PreviousYearCount': previous_count,
                    'YoYGrowthPct': growth_pct
                })
    
    # Convert to DataFrame
    yoy_df = pd.DataFrame(yoy_growth)
    
    # Create a date column for plotting (using the 1st of each month)
    yoy_df['Date'] = pd.to_datetime(yoy_df['Year'].astype(str) + '-' + yoy_df['Month'].astype(str) + '-01')
    
    return yoy_df

def plot_yoy_growth(yoy_df):
    # Set up the plot
    plt.figure(figsize=(14, 8))
    
    # Create a bar plot
    bars = plt.bar(yoy_df['Date'], yoy_df['YoYGrowthPct'], color=yoy_df['YoYGrowthPct'].apply(
        lambda x: 'green' if x >= 0 else 'red'))
    
    # Add a horizontal line at 0%
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    
    # Add labels and title
    plt.title('Year-on-Year Monthly Crime Growth (%)', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('YoY Growth (%)', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    
    # Format x-axis to show dates properly
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
    plt.xticks(rotation=45)
    
    # Add average line
    avg_growth = yoy_df['YoYGrowthPct'].mean()
    plt.axhline(y=avg_growth, color='blue', linestyle='--', alpha=0.7, 
                label=f'Avg Growth: {avg_growth:.2f}%')
    plt.legend()
    
    # Add data labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., 
                 height + (5 if height >= 0 else -10),
                 f'{height:.1f}%', 
                 ha='center', va='bottom' if height >= 0 else 'top', 
                 rotation=90, fontsize=8)
    
    plt.tight_layout()
    
    return plt

# Generate tabular data for detailed analysis
def generate_yoy_summary(yoy_df):
    # Pivot the data to create a table with months as rows and years as columns
    pivot_df = yoy_df.pivot_table(
        index=['Month', 'MonthName'], 
        columns='Year', 
        values='YoYGrowthPct', 
        aggfunc='first'
    ).reset_index()
    
    # Sort by month
    pivot_df = pivot_df.sort_values('Month')
    
    # Remove the Month column and keep only MonthName
    pivot_df = pivot_df.drop('Month', axis=1)
    
    # Rename columns for clarity
    pivot_df = pivot_df.rename(columns={'MonthName': 'Month'})
    
    # Format the percentage values
    for col in pivot_df.columns:
        if col != 'Month':
            pivot_df[col] = pivot_df[col].apply(lambda x: f'{x:.2f}%' if pd.notnull(x) else 'N/A')
    
    return pivot_df
```

```python
yoy_df = calculate_yoy_growth(df)
plot = plot_yoy_growth(yoy_df)
summary_table = generate_yoy_summary(yoy_df)

print(summary_table)
plot.show()
```

```python
def create_mirror_yoy_chart(df):
    # Get monthly counts
    monthly_counts = df[df['DATE OCC'].dt.year < 2025].groupby(df['DATE OCC'].dt.strftime('%Y-%m'))['DR_NO'].count()
    
    # Convert to DataFrame for easier manipulation
    monthly_df = monthly_counts.reset_index()
    monthly_df.columns = ['Date', 'Crime_Count']
    
    # Convert to datetime for proper sorting and year-month extraction
    monthly_df['Date'] = pd.to_datetime(monthly_df['Date'] + '-01')
    monthly_df = monthly_df.sort_values('Date')
    
    # Extract year and month for comparison
    monthly_df['Year'] = monthly_df['Date'].dt.year
    monthly_df['Month'] = monthly_df['Date'].dt.month
    
    # Pivot data to get counts by year and month
    pivot_df = monthly_df.pivot_table(index='Month', columns='Year', values='Crime_Count', aggfunc='first')
    
    # Calculate year-over-year growth
    for year in sorted(monthly_df['Year'].unique())[1:]:  # Skip the first year
        pivot_df[f'Growth_{year}'] = ((pivot_df[year] - pivot_df[year-1]) / pivot_df[year-1]) * 100
    
    # Create figure and axes
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Get years for which we have growth data
    growth_years = sorted(monthly_df['Year'].unique())[1:]
    
    # Define colors for different years (blue gradient)
    colors = plt.cm.Blues(np.linspace(0.4, 0.8, len(growth_years)))
    
    # Define month names for x-axis
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    
    # Width of each bar
    bar_width = 0.8 / len(growth_years)
    
    # Plot bars for each year's growth
    for i, year in enumerate(growth_years):
        growth_col = f'Growth_{year}'
        
        # Filter out missing values
        valid_months = pivot_df[growth_col].notna()
        plot_months = pivot_df.index[valid_months]
        plot_values = pivot_df.loc[valid_months, growth_col]
        
        # Calculate positions for bars
        positions = np.arange(1, 13) - 0.4 + (i * bar_width)
        positions = positions[valid_months]
        
        # Plot positive values
        positive = plot_values >= 0
        if any(positive):
            ax.bar(positions[positive], 
                   plot_values[positive], 
                   width=bar_width, 
                   color=colors[i], 
                   label=f'{year}')
            
            # Add raw count labels on top of positive bars
            for j, (month, value, pos) in enumerate(zip(
                plot_months[positive], 
                plot_values[positive], 
                positions[positive]
            )):
                # Get the actual count for this month/year
                count = pivot_df.loc[month, year]
                ax.text(pos, value + 2, f'{int(count):,}', 
                       ha='center', va='bottom', fontsize=8, rotation=90)
        
        # Plot negative values
        negative = plot_values < 0
        if any(negative):
            ax.bar(positions[negative], 
                   plot_values[negative], 
                   width=bar_width, 
                   color=colors[i], 
                   label='_nolegend_')  # Don't duplicate in legend
            
            # Add raw count labels below negative bars
            for j, (month, value, pos) in enumerate(zip(
                plot_months[negative], 
                plot_values[negative], 
                positions[negative]
            )):
                # Get the actual count for this month/year
                count = pivot_df.loc[month, year]
                ax.text(pos, value - 2, f'{int(count):,}', 
                       ha='center', va='top', fontsize=8, rotation=90)
    
    # Format y-axis as percentage
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.0f}%'))
    
    # Add a horizontal line at 0%
    ax.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    
    # Set x-axis labels to month names
    ax.set_xticks(np.arange(1, 13))
    ax.set_xticklabels(months)
    
    # Set labels and title
    ax.set_xlabel('Month', fontsize=12)
    ax.set_ylabel('Year-over-Year Growth (%)', fontsize=12)
    ax.set_title('Monthly Crime Count Year-over-Year Growth', fontsize=16)
    
    # Add grid lines for readability
    ax.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Add a legend
    ax.legend(title='Year', loc='upper right')
    
    # Add text explaining the chart
    fig.text(0.5, 0.01, 
             "Bar height shows YoY growth percentage. Numbers indicate actual crime count.", 
             ha='center', fontsize=10)
    
    plt.tight_layout(rect=[0, 0.03, 1, 1])  # Make room for the note at bottom
    
    return plt

plt = create_mirror_yoy_chart(df)
plt.show()
```

```python
def create_mirror_yoy_chart(df):
    monthly_counts = df[df['DATE OCC'].dt.year < 2025].groupby(df['DATE OCC'].dt.strftime('%Y-%m'))['DR_NO'].count()
    monthly_df = monthly_counts.reset_index()
    monthly_df.columns = ['Date', 'Crime_Count']
    monthly_df['Date'] = pd.to_datetime(monthly_df['Date'] + '-01')
    monthly_df = monthly_df.sort_values('Date')
    monthly_df['Year'] = monthly_df['Date'].dt.year
    monthly_df['Month'] = monthly_df['Date'].dt.month
    
    pivot_df = monthly_df.pivot_table(index='Month', columns='Year', values='Crime_Count', aggfunc='first')
    
    year_pairs = [(2021, 2022), (2022, 2023), (2023, 2024)]
    
    for base_year, compare_year in year_pairs:
        if base_year in pivot_df.columns and compare_year in pivot_df.columns:
            pivot_df[f'Growth_{compare_year}'] = ((pivot_df[compare_year] - pivot_df[base_year]) / pivot_df[base_year]) * 100
    
    fig, ax = plt.subplots(figsize=(14, 8))
    
    growth_years = [pair[1] for pair in year_pairs if pair[0] in pivot_df.columns and pair[1] in pivot_df.columns]
    
    pastel_green = '#a8e6cf'  
    pastel_red = '#ffb3ba'    
    
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    
    bar_width = 0.8 / len(growth_years)
    
    # Plot bars for each year's growth
    for i, year in enumerate(growth_years):
        growth_col = f'Growth_{year}'
        
        # Filter out missing values
        valid_months = pivot_df[growth_col].notna()
        plot_months = pivot_df.index[valid_months]
        plot_values = pivot_df.loc[valid_months, growth_col]
        
        # Calculate positions for bars
        positions = np.arange(1, 13) - 0.4 + (i * bar_width)
        positions = positions[valid_months]
        
        # Plot positive values with pastel green
        positive = plot_values >= 0
        if any(positive):
            ax.bar(positions[positive], 
                   plot_values[positive], 
                   width=bar_width, 
                   color=pastel_green, 
                   label=f'{year-1}-{year}' if i == 0 else f'{year-1}-{year}')  # Update legend label
            
            # Add raw count labels on top of positive bars (with more space)
            for j, (month, value, pos) in enumerate(zip(
                plot_months[positive], 
                plot_values[positive], 
                positions[positive]
            )):
                # Get the actual count for this month/year
                count = pivot_df.loc[month, year]
                # Add more space (value + 3 instead of value + 2)
                ax.text(pos, value + 3, f'{int(count):,}', 
                       ha='center', va='bottom', fontsize=8, rotation=90)
        
        # Plot negative values with pastel red
        negative = plot_values < 0
        if any(negative):
            ax.bar(positions[negative], 
                   plot_values[negative], 
                   width=bar_width, 
                   color=pastel_red, 
                   label=f'_nolegend_')  # Only add to legend once
            
            # Add raw count labels below negative bars (with more space)
            for j, (month, value, pos) in enumerate(zip(
                plot_months[negative], 
                plot_values[negative], 
                positions[negative]
            )):
                # Get the actual count for this month/year
                count = pivot_df.loc[month, year]
                # Add more space (value - 5 instead of value - 2)
                ax.text(pos, value - 5, f'{int(count):,}', 
                       ha='center', va='top', fontsize=8, rotation=90)
    
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.0f}%'))    
    ax.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax.set_xticks(np.arange(1, 13))
    ax.set_xticklabels(months)
    ax.tick_params(axis='both', length=0)
    ax.set_xlabel("")
    ax.set_ylabel("")
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title('') 
    
    fig.text(0.07, 0.95, 'Monthly Crime Count Year-over-Year Growth', 
             ha='left', fontsize=20, fontweight='bold')
    
    fig.text(0.07, 0.91, 'Bar height shows YoY growth percentage. Numbers indicate actual crime count.',
             ha='left', fontsize=12,alpha = 0.5)
        
    ax.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Expand the y-axis limits to prevent text interference with spines
    y_min, y_max = ax.get_ylim()
    ax.set_ylim(y_min - 10, y_max + 10)  # Add more padding
    
    # Create a custom legend for years and colors
    handles, labels = ax.get_legend_handles_labels()
    
    # Add custom patches for color legend
    green_patch = mpatches.Patch(color=pastel_green, label='Positive Growth')
    red_patch = mpatches.Patch(color=pastel_red, label='Negative Growth')
    
    # Add year handles and color handles to legend
    all_handles = handles + [green_patch, red_patch]
    all_labels = labels + ['Positive Growth', 'Negative Growth']
    
#     ax.legend(all_handles, all_labels, 
#               title='Legend', 
#               loc='upper left',
#               bbox_to_anchor=(1.01, 1),  # Position outside the right edge of the plot
#               borderaxespad=0)
    
    # Add more padding to the right to accommodate the legend
    plt.tight_layout(rect=[0, 0, 0.85, 0.88])  # Adjust right margin for legend and top margin for title
    
    return plt

plt = create_mirror_yoy_chart(df)
plt.show()
```

# Weapons

```python
def categorize_weapon(weapon):
    if pd.isna(weapon):
        return "No Weapon/Unknown"
    
    # Firearms
    elif any(w in weapon for w in ["HAND GUN", "SEMI-AUTOMATIC PISTOL", "UNKNOWN FIREARM", 
                                   "REVOLVER", "OTHER FIREARM", "RIFLE", "SHOTGUN", 
                                   "SEMI-AUTOMATIC RIFLE", "SAWED OFF RIFLE/SHOTGUN", 
                                   "RELIC FIREARM", "ANTIQUE FIREARM"]):
        return "Firearms"
    
    # Assault Weapons
    elif any(w in weapon for w in ["ASSAULT WEAPON", "UZI", "AK47", "AUTOMATIC WEAPON", 
                                   "SUB-MACHINE GUN", "SEMIAUTOMATIC ASSAULT", 
                                   "HECKLER & KOCH", "M-14", "MAC-10", "MAC-11"]):
        return "Assault Weapons"
    
    # Simulated/Toy/Non-lethal Firearms
    elif any(w in weapon for w in ["AIR PISTOL", "BB GUN", "SIMULATED GUN", "TOY GUN", 
                                   "STARTER PISTOL", "STUN GUN"]):
        return "Simulated/Non-lethal Firearms"
    
    # Bladed Weapons
    elif any(w in weapon for w in ["KNIFE", "BLADE", "MACHETE", "CUTTING INSTRUMENT", 
                                   "SCISSORS", "SWITCH BLADE", "SWORD", "RAZOR", 
                                   "DIRK/DAGGER", "ICE PICK", "CLEAVER", "BOWIE"]):
        return "Bladed Weapons"
    
    # Blunt Instruments
    elif any(w in weapon for w in ["PIPE", "STICK", "CLUB", "BAT", "BLUNT INSTRUMENT", 
                                   "HAMMER", "CONCRETE", "BRICK", "BOARD", "TIRE IRON", 
                                   "BLACKJACK", "FIXED OBJECT"]):
        return "Blunt Instruments"
    
    # Thrown/Projectile Weapons
    elif any(w in weapon for w in ["ROCK", "THROWN OBJECT", "BOTTLE", "GLASS", "BOW AND ARROW"]):
        return "Thrown/Projectile Weapons"
    
    # Bodily Force
    elif any(w in weapon for w in ["STRONG-ARM", "BODILY FORCE", "PHYSICAL PRESENCE", 
                                   "MARTIAL ARTS"]):
        return "Bodily Force"
    
    # Chemical/Biological
    elif any(w in weapon for w in ["MACE", "PEPPER SPRAY", "CHEMICAL", "POISON", 
                                   "SCALDING", "LIQUOR", "DRUGS", "SYRINGE"]):
        return "Chemical/Biological"
    
    # Environmental/Misc
    elif any(w in weapon for w in ["VEHICLE", "FIRE", "AXE", "BELT", "CHAIN", 
                                   "BRASS KNUCKLES", "ROPE", "LIGATURE", "DOG", "ANIMAL"]):
        return "Environmental/Misc"
    
    # Explosive
    elif any(w in weapon for w in ["BOMB", "EXPLOSIVE"]):
        return "Explosive"
    
    # Non-Physical Threats
    elif any(w in weapon for w in ["VERBAL THREAT", "DEMAND NOTE"]):
        return "Non-Physical Threats"
    
    # Unclassified
    else:
        return "Other/Unclassified"
```

```python
df['Weapon Category'] = df['Weapon Desc'].apply(categorize_weapon)
weapon_use_counts = df[df['Weapon Desc'].notna()].groupby('Weapon Category')['DR_NO'].count().reset_index().rename(columns = {'DR_NO':'Count'})
```

```python
status_counts = df['Status Desc'].value_counts()
status_percentages = (status_counts / status_counts.sum() * 100).reset_index()
status_percentages.columns = ['Status', 'Percentage']

plt.figure(figsize=(10, 6))
sns.barplot(x='Percentage', y='Status', data=status_percentages,color = 'gray')
plt.title('Percentage of Crimes by Status')
plt.xlabel('Percentage (%)')
plt.ylabel('Status')
plt.tight_layout()
plt.show()

# plt.figure(figsize=(10, 6))
# plt.pie(status_percentages['Percentage'], labels=status_percentages['Status'], 
#         autopct='%1.1f%%', startangle=90)
# plt.axis('equal')
# plt.title('Distribution of Crime Outcomes')
# plt.tight_layout()
# plt.show()
```

```python
# Calculate clearance rate by crime type
# Assuming "Adult Arrest", "Juvenile Arrest", etc. indicate cleared cases

cleared_statuses = ['Adult Arrest', 'Juvenile Arrest', 'Adult Other', 'Juvenile Other']  # Adjust as needed
df['Cleared'] = df['Status Desc'].apply(lambda x: 1 if x in cleared_statuses else 0)
crime_clearance = df.groupby('Crm Cd Desc').agg(
    total_crimes=('DR_NO', 'count'),
    cleared_crimes=('Cleared', 'sum')
).reset_index()
crime_clearance['clearance_rate'] = (crime_clearance['cleared_crimes'] / 
                                    crime_clearance['total_crimes'] * 100).round(2)

top_crimes_by_clearance = crime_clearance.sort_values('total_crimes', ascending=False).head(10)
top_crimes_by_clearance = top_crimes_by_clearance.sort_values('clearance_rate',ascending = False)

plt.figure(figsize=(12, 10))
sns.barplot(x='clearance_rate', y='Crm Cd Desc', data=top_crimes_by_clearance,color = 'gray')
plt.title('Clearance Rates for Top 10 Most Common Crime Types')
plt.xlabel('Clearance Rate (%)')
plt.ylabel('Crime Type')
plt.tight_layout()
plt.show()
```

```python
# Clearance rate by area
area_clearance = df.groupby('AREA NAME').agg(
    total_crimes=('DR_NO', 'count'),
    cleared_crimes=('Cleared', 'sum')
).reset_index()

area_clearance['clearance_rate'] = (area_clearance['cleared_crimes'] / 
                                   area_clearance['total_crimes'] * 100).round(2)

area_clearance = area_clearance.sort_values('clearance_rate')

plt.figure(figsize=(12, 8))
sns.barplot(x='clearance_rate', y='AREA NAME', data=area_clearance,color = 'gray')
plt.title('Crime Clearance Rates by LAPD Area')
plt.xlabel('Clearance Rate (%)')
plt.ylabel('Area')
plt.tight_layout()
plt.show()
```
