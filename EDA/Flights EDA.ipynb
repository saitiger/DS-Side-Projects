```python
import pandas as pd
```

```python
df = pd.read_csv(r'flights.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['arr_time'].isnull().sum()
```

```python
df.drop(df.loc[df['arr_time'].isnull()].index,inplace=True)
```

```python
df = df.reset_index()
```

```python
df.drop(df.loc[df['arr_delay'].isnull()].index,inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df['arr_delay'].isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df.dropna(axis=0,how='any',inplace=True)
```

```python
filt1 = df['dep_delay']<0
df.loc[filt1,'dep_delay']=0

filt2 = df['arr_delay']<0
df.loc[filt2,'arr_delay']=0
```

```python
check = df['dep_delay']<0
print("Departure Delay negative numbers : ",check.sum())

check2 = df['arr_delay']<0
print("Arrival Delay negative numbers : ", check2.sum())
```
