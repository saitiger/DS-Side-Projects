### Query 6.1 
Create a table/dataframe with the closing prices of 30 different stocks, with 10 from each of the caps

```python
import pandas as pd
from pylab import plot,show
from numpy import vstack,array
from numpy.random import rand
import numpy as np
from scipy.cluster.vq import kmeans,vq
from math import sqrt
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

from sklearn import metrics
from scipy.spatial.distance import cdist

```

```python
grp = pd.DataFrame()

ad = pd.read_csv('ADANIPORTS.csv')
grp["ADANIPORTS"] = ad['Close Price']

am = pd.read_csv('ASIANPAINT.csv')
grp["ASIANPAINT"] = am['Close Price']

ap = pd.read_csv('APOLLOTYRE.csv')
grp["APOLLO"] = ap['Close Price']

ash = pd.read_csv('ASHOKA.csv')
grp["ASHOKA"] = ash['Close Price']

axb = pd.read_csv('AXISBANK.csv')
grp["AXISBANK"] = axb['Close Price']

bj = pd.read_csv('BAJAJELEC.csv')
grp["BAJAJELEC"] = bj['Close Price']

bmb = pd.read_csv('BOMDYEING.csv')
grp["BOMDYEING"] = bmb['Close Price']

c = pd.read_csv('CASTROLIND.csv')
grp["CASTROLIND"] = c['Close Price']

ci = pd.read_csv('CIPLA.csv')
grp["CIPLA"] = ci['Close Price']

cp = pd.read_csv('CENTURYPLY.csv')
grp["CENTURYPLY"] = cp['Close Price']

fr = pd.read_csv('FORTIS.csv')
grp["FORTIS"] = fr['Close Price']

hd = pd.read_csv('HDFC.csv')
grp["HDFC"] = hd['Close Price']

i = pd.read_csv('IDBI.csv')
grp["IDBI"] = i['Close Price']

idf = pd.read_csv('IDFC.csv')
grp["IDFC"] = idf['Close Price']

inf = pd.read_csv('INFY.csv')
grp["INFY"] = inf['Close Price']

irc = pd.read_csv('IRCON.csv')
grp["ADANIPORTS"] = irc['Close Price']

itc = pd.read_csv('ITC.csv')
grp["ITC"] = itc['Close Price']

itd =pd.read_csv('ITDC.csv')
grp["ITDC"] = itd['Close Price']

jet = pd.read_csv('JETAIRWAYS.csv')
grp["JET"] = jet['Close Price']

lt = pd.read_csv('LT.csv')
grp["LT"] = lt['Close Price']

m = pd.read_csv('M&M.csv')
grp["M&M"] = m['Close Price']

mi = pd.read_csv('MINDTREE.csv')
grp["MINDTREE"] = mi['Close Price']

p = pd.read_csv('PNB.csv')
grp["PNB"] = p['Close Price']

rl = pd.read_csv('RELIANCE.csv')
grp["RELIANCE"] = rl['Close Price']

rel = pd.read_csv('RELINFRA.csv')
grp["RELINFRA"] = rel['Close Price']

sun = pd.read_csv('SUNPHARMA.csv')
grp["SUNPHARMA"] = sun['Close Price']

tv = pd.read_csv('SUNTV.csv')
grp["SUNTV"] = tv['Close Price']

tp = pd.read_csv('TATAPOWER.csv')
grp["TATAPOWER"] = tp['Close Price']

tcs = pd.read_csv('TCS.csv')
grp["TCS"] = tcs['Close Price']

vol = pd.read_csv('VOLTAS.csv')
grp["VOLTAS"] = vol['Close Price']

gK = pd.read_csv('GUJALKALI.csv')
grp["GOLD"] = gK['Close Price']

grp
```

```python
grp = grp.dropna()
returns = grp.astype(float).pct_change().mean() * 252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns['Volatility'] = grp.pct_change().astype(float).std() * sqrt(252)
returns
```

```python
# Finding the optimal value of k to apply to pass as the parameter in KMeans
data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T
X = data
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# computing K-Means with K = 3 (3 clusters)
centroids,_ = kmeans(X,3)
# assign each sample to a cluster
idx,_ = vq(data,centroids)
# some plotting using numpy's logical indexing
plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'oy',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',
     data[idx==4,0],data[idx==4,1],'om')
plot(centroids[:,0],centroids[:,1],'sg',markersize=3)
```
