Query 1.2
Calculate the maximum, minimum and mean price for the last 90 days. (price=Closing Price unless stated otherwise)

```python
dataset.describe()
#Since the all the values of Date are unique there is no need to do any alterations in the data. 
```

```python

#Changing the datatype from string to float to make calculations using functions valid and removing , from the values

dataset.iloc[:,1:6]= dataset.iloc[:,1:6].replace(to_replace ='[,M]', value = '', regex = True)
for j in range (1,6):
    dataset.iloc[:,j]=dataset.iloc[:,j].astype(float)

# Selecting the 90 rows which are the last 90 days since the data is arranged in reverse chronological order 
dt1 = dataset.iloc[:91,:]


# Singling out the Close Price column to find the min,max and mean 
price = np.array(dt1['Close Price'])

print(price)

# Max
print("Maximum Price : " , price.max())

# Min
print("Minimum Price : " , price.min())

# Mean 
print("Mean is : " , price.mean())
```

```python
# Calculating VWAP 

dt1 = dataset.copy() # Created a copy so that original database is not tinkered with

# Extracting Date information from the dataset
ex_date = dataset['Date'] 
dt1['Month'] = ex_date.dt.month 
dt1['Year'] = ex_date.dt.year
dt1['New Price']=dataset['Close Price']*dataset['Total Traded Quantity']

# Grouped the dataset on the basis of month and year 

vwap = dt1.groupby(['Month','Year']).sum()['New Price']/dt1.groupby(['Month','Year']).sum()['Total Traded Quantity']
print(vwap)
```

```python
dataset['Day_Perc_Change'] = dataset['Close Price'].pct_change()*100
dataset = dataset.drop(dataset.index[0])
```

QUERY 1.7

```python
def cases(num):
    print(num)
    if num >= -0.5 and num <= 0.5 : return 'Slight or No change'
    elif num >= 0.5 and num <= 1 : return 'Slight positive'
    elif num >= -1 and num <= -0.5 : return 'Slight negative'
    elif num >= 1 and num <= 3 : return 'Positive'
    elif num >= -3  and num <= -1 : return 'Negative'
    elif num >= 3 and num <= 7 : return 'Among top gainers'
    elif num >= -7  and num <= -3 : return 'Among top losers'
    elif num > 7 : return 'Bull run'
    elif num < -7 : return 'Bear drop'

dataset["Trend"] = dataset.Day_Perc_Change.apply(cases)
```

```python
print(dataset[['Trend', 'Total Traded Quantity']].groupby(['Trend'], as_index=False).mean().sort_values(by='Total Traded Quantity', ascending=False))
```

QUERY 1.8

```python
# Finding the average and median

Mean = dataset.groupby('Trend').mean()
Median = dataset.groupby('Trend').median()
print('Mean is :' , Mean)
print('Median is : ' , Median)
```
